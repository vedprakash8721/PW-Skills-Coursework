# Q.No. 1: Compare and contrast NLTK and spaCy in terms of features, ease of use, and performance.

Answer:

NLTK (Natural Language Toolkit) and spaCy are popular Python libraries for Natural Language Processing (NLP), but they serve different purposes.

NLTK is mainly designed for learning and research. It provides a wide range of NLP tools such as tokenization, stemming, parsing, and text classification. It is beginner-friendly and offers many educational resources.

spaCy is designed for building real-world NLP applications. It is faster than NLTK and provides efficient features like tokenization, lemmatization, part-of-speech tagging, dependency parsing, and named entity recognition.

Overall, NLTK is suitable for academic learning and experimentation, while spaCy is preferred for production-level applications due to its speed and better performance.

# Q.No. 2: What is TextBlob and how does it simplify common NLP tasks like sentiment analysis and translation?

Answer:

TextBlob is a simple Python library built on top of NLTK that makes common NLP tasks easy to perform. It provides a user-friendly interface for beginners and requires very little code.

Using TextBlob, developers can perform tasks such as sentiment analysis, translation, spelling correction, noun phrase extraction, tokenization, and part-of-speech tagging.

For sentiment analysis, TextBlob returns two values: polarity, which measures whether the text is positive or negative, and subjectivity, which measures whether the text expresses opinions or facts.

Overall, TextBlob is an excellent library for beginners who want to perform basic NLP tasks quickly without implementing complex algorithms.

# Q.No. 4: Describe the architecture and functioning of a Recurrent Neural Network (RNN).

Answer:

A Recurrent Neural Network (RNN) is a deep learning model designed to process sequential data such as text, speech, and time-series data. Unlike traditional neural networks, an RNN has connections that allow information from previous time steps to be used when processing the current input.

An RNN processes one word or sequence element at a time while maintaining a hidden state that stores information from earlier inputs. This enables the network to understand the context of a sequence.

RNNs are commonly used in language translation, speech recognition, sentiment analysis, and text generation. However, they may suffer from the vanishing gradient problem when handling long sequences.

# Q.No. 5: What is the key difference between LSTM and GRU networks in NLP applications?

Answer:

LSTM (Long Short-Term Memory) and GRU (Gated Recurrent Unit) are advanced versions of RNNs that solve the vanishing gradient problem.

The main difference is that LSTM uses three gates (input, forget, and output) along with a cell state to remember long-term information. GRU uses only two gates (reset and update) and combines the hidden state with the memory, making it simpler.

LSTM is generally better for complex tasks involving long sequences, while GRU trains faster and requires fewer parameters. Both models are widely used in text classification, language translation, and speech recognition.

# Q.No. 6: Write a Python program using TextBlob to perform sentiment analysis.

```python
from textblob import TextBlob

text = """
I had a great experience using the new mobile banking app.
The interface is intuitive, and customer support was quick to resolve my issue.
However, the app did crash once during a transaction, which was frustrating.
"""

blob = TextBlob(text)

print("Polarity:", blob.sentiment.polarity)
print("Subjectivity:", blob.sentiment.subjectivity)
```

# Q.No. 7: Perform string tokenization and frequency distribution using Python and NLTK.

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

nltk.download('punkt')

text = """
Natural Language Processing (NLP) is a fascinating field that combines linguistics,
computer science, and artificial intelligence. It enables machines to understand,
interpret, and generate human language. Applications of NLP include chatbots,
sentiment analysis, and machine translation. As technology advances, the role of NLP
in modern solutions is becoming increasingly critical.
"""

tokens = word_tokenize(text)

print("Tokens:")
print(tokens)

freq = FreqDist(tokens)

print("\nWord Frequency:")
for word, count in freq.items():
    print(word, ":", count)
```

# Q.No. 8: Implement a basic LSTM model in Keras for text classification.

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

texts = [
    "I love this project",
    "This is an amazing experience",
    "I hate waiting in line",
    "This is the worst service",
    "Absolutely fantastic!"
]

labels = [1, 1, 0, 0, 1]

tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)

sequences = tokenizer.texts_to_sequences(texts)
X = pad_sequences(sequences, maxlen=5)

model = Sequential([
    Embedding(input_dim=30, output_dim=16, input_length=5),
    LSTM(32),
    Dense(1, activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.fit(X, labels, epochs=20, verbose=1)

loss, accuracy = model.evaluate(X, labels, verbose=0)

print("Accuracy:", accuracy)
```

# Q.No. 9: Using spaCy, build a simple NLP pipeline.

```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = """
Homi Jehangir Bhaba was an Indian nuclear physicist who played a key role in the
development of India’s atomic energy program. He was the founding director of the
Tata Institute of Fundamental Research (TIFR) and was instrumental in establishing
the Atomic Energy Commission of India.
"""

doc = nlp(text)

print("Tokens and Lemmas:")
for token in doc:
    print(token.text, "->", token.lemma_)

print("\nNamed Entities:")
for ent in doc.ents:
    print(ent.text, "-", ent.label_)
```
