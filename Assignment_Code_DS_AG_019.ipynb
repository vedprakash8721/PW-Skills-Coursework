# Question 1: What is Deep Learning? Briefly describe how it evolved and how it differs from traditional machine learning.

Answer:

Deep Learning is a branch of Artificial Intelligence (AI) and Machine Learning that uses artificial neural networks with multiple hidden layers to learn patterns from large amounts of data. It is capable of automatically extracting important features from raw data without requiring much manual feature engineering.

Deep Learning evolved from Artificial Neural Networks (ANNs). Early neural networks were limited because of low computing power and small datasets. With the advancement of powerful GPUs, large datasets, and improved algorithms, Deep Learning became highly successful in solving complex problems such as image recognition, speech recognition, natural language processing, and autonomous driving.

The main difference between Deep Learning and traditional Machine Learning is that traditional Machine Learning usually requires manual feature extraction, while Deep Learning automatically learns features from the data. Deep Learning also performs better on large datasets but requires more computational resources and training time.

# Question 2: Explain the basic architecture and functioning of a Perceptron. What are its limitations?

Answer:

A Perceptron is the simplest type of artificial neural network and is mainly used for binary classification problems. It consists of three main components: the input layer, weights, and an output neuron. Each input is multiplied by its corresponding weight, and all weighted inputs are added together along with a bias. The result is then passed through an activation function to produce the final output.

The perceptron learns by adjusting its weights based on the prediction error during training. This process continues until the model achieves satisfactory performance.

However, the perceptron has some limitations. It can only solve linearly separable problems and cannot solve complex problems like the XOR gate. It also has limited learning capability because it contains only a single layer and cannot capture complex relationships in data.

# Question 3: Describe the purpose of activation function in neural networks. Compare Sigmoid, ReLU, and Tanh functions.

Answer:

An activation function determines whether a neuron should be activated or not. It introduces non-linearity into the neural network, allowing the model to learn complex patterns and relationships in the data. Without an activation function, a neural network would behave like a simple linear model regardless of the number of layers.

Sigmoid Function:

Produces output values between 0 and 1.
Commonly used for binary classification.
Can suffer from the vanishing gradient problem.

ReLU (Rectified Linear Unit):

Produces the input value if it is positive; otherwise, it outputs zero.
It is computationally efficient and trains faster.
It is the most commonly used activation function in deep learning.

Tanh Function:

Produces output values between -1 and 1.
It is centered around zero, making learning more stable than Sigmoid.
It can also suffer from the vanishing gradient problem.

Among these, ReLU is generally preferred for hidden layers because it provides faster convergence and better performance.

# Question 4: What is the difference between Loss function and Cost function in neural networks? Provide examples.

Answer:

A Loss Function measures the error for a single training example. It calculates the difference between the predicted output and the actual output for one data point. The optimizer uses this value to update the model's parameters during training.

A Cost Function is the average of the loss values over the entire training dataset. It provides an overall measure of how well the model is performing on all training examples.

Examples of Loss Functions:

Mean Squared Error (MSE)
Binary Cross-Entropy
Categorical Cross-Entropy

Example of Cost Function:
If the loss values for five training samples are calculated, their average becomes the cost function value.

In simple words, Loss is calculated for one sample, whereas Cost is calculated for the complete dataset.

# Question 5: What is the role of optimizers in neural networks? Compare Gradient Descent, Adam, and RMSprop.

Answer:

Optimizers are algorithms that help a neural network learn by updating the model's weights and biases to minimize the loss function. Their main goal is to improve the model's accuracy while reducing prediction errors during training.

Gradient Descent:
Gradient Descent updates the model parameters using the entire training dataset. It is simple and easy to understand but can be slow for large datasets.

RMSprop:
RMSprop adjusts the learning rate automatically for each parameter. It performs well on complex neural networks and helps achieve faster convergence compared to standard Gradient Descent.

Adam (Adaptive Moment Estimation):
Adam combines the advantages of Momentum and RMSprop. It adapts the learning rate during training and usually converges faster than other optimizers. Adam is one of the most widely used optimizers in deep learning because of its efficiency and stability.

Overall, Adam is generally considered the best choice for most deep learning applications due to its fast convergence and reliable performance.

# Question 6: Write a Python program to implement a single-layer perceptron from scratch using NumPy to solve the logical AND gate.

```python
import numpy as np

# Input dataset
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

# Target output
y = np.array([0, 0, 0, 1])

# Initialize weights and bias
weights = np.zeros(2)
bias = 0
learning_rate = 0.1

# Perceptron training
for epoch in range(10):
    for i in range(len(X)):
        linear_output = np.dot(X[i], weights) + bias
        prediction = 1 if linear_output >= 0 else 0

        error = y[i] - prediction

        weights += learning_rate * error * X[i]
        bias += learning_rate * error

# Testing
print("Predictions:")
for x in X:
    output = 1 if np.dot(x, weights) + bias >= 0 else 0
    print(f"{x} -> {output}")
```

# Question 7: Implement and visualize Sigmoid, ReLU, and Tanh activation functions using Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 200)

sigmoid = 1 / (1 + np.exp(-x))
relu = np.maximum(0, x)
tanh = np.tanh(x)

plt.figure(figsize=(10,6))

plt.plot(x, sigmoid, label='Sigmoid')
plt.plot(x, relu, label='ReLU')
plt.plot(x, tanh, label='Tanh')

plt.title("Activation Functions")
plt.xlabel("Input")
plt.ylabel("Output")
plt.legend()

plt.grid(True)
plt.show()
```

# Question 8: Use Keras to build and train a simple multilayer neural network on the MNIST digits dataset. Print the training accuracy.

```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten

(X_train, y_train), (X_test, y_test) = mnist.load_data()

X_train = X_train / 255.0
X_test = X_test / 255.0

model = Sequential([
    Flatten(input_shape=(28,28)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

history = model.fit(X_train, y_train, epochs=5)

print("Training Accuracy:", history.history['accuracy'][-1])
```

# Question 9: Visualize the loss and accuracy curves for a neural network model trained on the Fashion MNIST dataset. Interpret the training behavior.


```python
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten

(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()

X_train = X_train / 255.0

model = Sequential([
    Flatten(input_shape=(28,28)),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

history = model.fit(X_train, y_train, epochs=5)

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(history.history['loss'])
plt.title("Loss Curve")
plt.xlabel("Epoch")
plt.ylabel("Loss")

plt.subplot(1,2,2)
plt.plot(history.history['accuracy'])
plt.title("Accuracy Curve")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")

plt.show()
```
