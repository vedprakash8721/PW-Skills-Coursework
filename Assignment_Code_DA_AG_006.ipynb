# Q.No. 1: What is a random variable in probability theory?

Answer:

A random variable is a variable whose value depends on the outcome of a random experiment. It assigns a numerical value to each possible outcome of an event. Random variables help describe uncertain events mathematically and are widely used in probability and statistics.

For example, when rolling a dice, the number obtained (1 to 6) is a random variable because the outcome is uncertain before the dice is rolled.

---
# Q.No. 2: What are the types of random variables?

Answer:

There are two main types of random variables:

Discrete Random Variable: It takes a finite or countable number of values. For example, the number of students in a class or the number shown on a dice.
Continuous Random Variable: It can take any value within a given range. For example, a person's height, weight, or temperature.

Discrete random variables are represented using probability mass functions (PMF), while continuous random variables are represented using probability density functions (PDF).

# Q.No. 3: Explain the difference between discrete and continuous distributions.

Answer:

A discrete distribution describes the probabilities of a discrete random variable, which takes specific countable values. Examples include the Binomial and Poisson distributions.

A continuous distribution describes the probabilities of a continuous random variable, which can take any value within a range. Examples include the Normal and Exponential distributions.

The main difference is that discrete distributions deal with countable values, while continuous distributions deal with measurable values. In continuous distributions, the probability of a single exact value is zero, and probabilities are calculated over intervals.

# Q.No. 4: What is a binomial distribution, and how is it used in probability?

Answer:

The Binomial Distribution is a probability distribution that models the number of successes in a fixed number of independent trials. Each trial has only two possible outcomes: success or failure, and the probability of success remains constant.

For example, if a coin is tossed 10 times, the binomial distribution can be used to calculate the probability of getting exactly 6 heads.

It is widely used in quality control, medical research, surveys, and other situations involving repeated experiments with two possible outcomes.

# Q.No. 5: What is the standard normal distribution, and why is it important?

Answer:

The Standard Normal Distribution is a normal distribution with a mean of 0 and a standard deviation of 1. It is represented by the symbol Z.

It is important because it allows different datasets to be compared using Z-scores. Many statistical methods, such as hypothesis testing and confidence intervals, are based on the standard normal distribution. It also helps calculate probabilities and identify unusual observations in a dataset.

# Q.No. 6: What is the Central Limit Theorem (CLT), and why is it critical in statistics?

Answer:

The Central Limit Theorem (CLT) states that when the sample size is sufficiently large, the distribution of sample means becomes approximately normal, regardless of the original population distribution.

The CLT is important because it allows statisticians to make reliable conclusions about a population using sample data. It forms the foundation of many statistical techniques, including confidence intervals and hypothesis testing. It is widely used in research, business, healthcare, and data analysis to estimate population parameters accurately.

# Q.No. 7: What is the significance of confidence intervals in statistical analysis?

Answer:

A confidence interval is a range of values that is likely to contain the true population parameter with a certain confidence level, such as 95%.

Confidence intervals help measure the reliability of sample estimates. Instead of providing only a single value, they show the possible range in which the actual population value may lie. A narrower confidence interval indicates a more precise estimate, while a wider interval indicates greater uncertainty.

Confidence intervals are commonly used in surveys, experiments, and business analytics for better decision-making.

# Q.No. 8: What is the concept of expected value in a probability distribution?

Answer:

The expected value is the average or long-term value of a random variable after many repeated experiments. It represents the weighted average of all possible outcomes based on their probabilities.

For a discrete random variable, the expected value is calculated by multiplying each outcome by its probability and then adding the results.

Expected value is widely used in statistics, economics, finance, and machine learning to predict average outcomes and support decision-making. It provides a measure of the central tendency of a probability distribution.

# Q.No. 9: Write a Python program to generate 1000 random numbers from a normal distribution with mean = 50 and standard deviation = 5. Compute its mean and standard deviation using NumPy, and draw a histogram to visualize the distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 random numbers
data = np.random.normal(loc=50, scale=5, size=1000)

# Compute mean and standard deviation
mean = np.mean(data)
std = np.std(data)

print("Mean:", mean)
print("Standard Deviation:", std)

# Plot histogram
plt.hist(data, bins=30, edgecolor='black')
plt.title("Normal Distribution")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

# Q.No. 10: You are working as a data analyst for a retail company...

Answer:

The Central Limit Theorem can be used to estimate the average daily sales because the sample mean follows an approximately normal distribution. First, we calculate the sample mean and standard deviation. Then, we compute the standard error and use the Z-value (1.96 for 95% confidence) to calculate the confidence interval. This interval provides the range within which the true average daily sales is expected to lie with 95% confidence.

```python
import numpy as np
from scipy import stats

daily_sales = [
    220, 245, 210, 265, 230,
    250, 260, 275, 240, 255,
    235, 260, 245, 250, 225,
    270, 265, 255, 250, 260
]

mean = np.mean(daily_sales)
std = np.std(daily_sales, ddof=1)
n = len(daily_sales)

confidence = 0.95
margin = stats.t.ppf((1 + confidence) / 2, n - 1) * (std / np.sqrt(n))

lower = mean - margin
upper = mean + margin

print("Mean Sales:", mean)
print("95% Confidence Interval:")
print(f"({lower:.2f}, {upper:.2f})")
```
