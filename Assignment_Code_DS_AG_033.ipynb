# Q.No. 1: What is a Vector Database (VectorDB) and how is it different from traditional databases?

Answer:

A Vector Database (VectorDB) is a specialized database designed to store and search vector embeddings generated by AI and machine learning models. These embeddings represent the semantic meaning of data such as text, images, audio, or videos.

Unlike traditional databases that search using exact values or keywords, VectorDB performs similarity searches by comparing vectors. This allows it to find data that is contextually similar rather than exactly matching.

Vector databases are widely used in AI applications such as semantic search, recommendation systems, chatbots, and Retrieval-Augmented Generation (RAG), where understanding the meaning of data is more important than exact keyword matching.

# Q.No. 2: Explain the various types of VectorDBs available and describe their suitability for different use cases.

Answer:

There are several popular Vector Databases used in AI and machine learning projects.

Chroma DB: Lightweight and easy to use for local AI applications and RAG systems.
Pinecone: A cloud-based managed vector database suitable for large-scale production applications.
Milvus: An open-source vector database designed for handling large datasets efficiently.
Weaviate: Supports semantic search and integrates well with machine learning models.
FAISS: Developed by Meta, it is mainly used for fast similarity search and clustering.

The choice of VectorDB depends on the project requirements. Chroma DB is ideal for beginners and local projects, Pinecone is suitable for cloud-based applications, Milvus handles large datasets efficiently, and FAISS is widely used for research and high-speed vector search.

# Q.No. 6: Install and set up Chroma DB, and insert sample vector data for semantic search.

```python
import chromadb

# Create Chroma client
client = chromadb.Client()

# Create Collection
collection = client.create_collection("sample_collection")

# Insert Sample Data
collection.add(
    documents=[
        "Artificial Intelligence",
        "Machine Learning",
        "Deep Learning"
    ],
    ids=["1", "2", "3"]
)

# Semantic Search
results = collection.query(
    query_texts=["AI"],
    n_results=2
)

print(results)
```

# Q.No. 8: Create a custom LLM using Ollama and Llama2, and run it locally for basic text prompts.

# Q.No. 9: Implement a basic RAG (Retrieval-Augmented Generation) system using Ollama with Llama3.

```python
import chromadb
import ollama

client = chromadb.Client()

collection = client.create_collection("rag_docs")

collection.add(
    documents=[
        "Python is a popular programming language.",
        "Machine Learning is a subset of Artificial Intelligence."
    ],
    ids=["1", "2"]
)

query = "What is Machine Learning?"

results = collection.query(
    query_texts=[query],
    n_results=1
)

context = results["documents"][0][0]

response = ollama.chat(
    model="llama3",
    messages=[
        {
            "role": "user",
            "content": f"Context: {context}\nQuestion: {query}"
        }
    ]
)

print(response["message"]["content"])
```

# Q.No. 10: A health-tech startup wants to build a chatbot that can answer user queries based on medical research articles. Propose and explain a solution using Hugging Face models for understanding, VectorDB for retrieval, and Ollama for generation.

Answer:

A medical chatbot can be built using Hugging Face, VectorDB, and Ollama. First, medical research articles are collected and converted into embeddings using a Hugging Face sentence embedding model. These embeddings are stored in a Vector Database such as Chroma DB for efficient semantic search.

When a user asks a medical question, the query is converted into an embedding and compared with the stored vectors. The most relevant research articles are retrieved from the VectorDB. This retrieved information is then provided as context to an Ollama model (such as Llama3), which generates a clear and informative response based on the retrieved documents.

This Retrieval-Augmented Generation (RAG) approach reduces hallucinations and provides more accurate answers based on trusted medical sources. The chatbot should also protect patient privacy, encrypt sensitive data, and clearly state that it provides informational support rather than professional medical advice. Human experts should review important medical recommendations before they are used in practice.
